```python
distances = []
for i in df_orders.values:
    dist = ((df_cities[(df_cities["city_from_name"] == i[1]) & (df_cities["city_to_name"] == i[4])]["distance"].values) or (df_cities[(df_cities["city_from_name"] == i[4]) & (df_cities["city_to_name"] == i[1])]["distance"].values))
    if (dist):
        distances.append(dist[0])
    else:
        distances.append(0)

df_orders["dist_to_hub"] = distances
```

```python
#delays per categoria
late = [0 for x in range(6)]
total = [0 for x in range(6)]

for i, id in enumerate(df_orders["product_id"]):
  total[df_orders[df_orders["product_id"] == id]["material_handling"].values[0]] += 1
  if df_orders[df_orders["product_id"] == id]["late_order"].values[0] == True:
    late[df_orders[df_orders["product_id"] == id]["material_handling"].values[0]] += 1

for i in range(6):
  if total[i] != 0:
    print(i, ":",late[i]/total[i]*100 , "%")
  else:
    print(i, ":", 0, "%")
```

```python
#delays per weight  
late = [0 for x in range(6)]
total = [0 for x in range(6)] 

for i, id in enumerate(df_orders["product_id"]):
  if df_orders[df_orders["product_id"] == id]["weight"].values[0] < 500:
    category_weight = 0
  elif df_orders[df_orders["product_id"] == id]["weight"].values[0] < 1000:
    category_weight = 1
  elif df_orders[df_orders["product_id"] == id]["weight"].values[0] < 1500:
    category_weight = 2
  elif df_orders[df_orders["product_id"] == id]["weight"].values[0] < 2000:
    category_weight = 3
  elif df_orders[df_orders["product_id"] == id]["weight"].values[0] < 2500:
    category_weight = 4
  else:
    category_weight = 5

  total[category_weight] += 1
  if df_orders[df_orders["product_id"] == id]["late_order"].values[0] == True:
    late[category_weight] += 1

for i in range(6):
  if total[i] != 0:
    print(i, ":",late[i]/total[i]*100 , "%")
  else:
    print(i, ":", 0, "%")
```

```python
#quins customs procedures
procedures = set()
for i, id in enumerate(df_orders["product_id"]):
  procedures.add(df_orders[df_orders["product_id"]== id]["customs_procedures"].values[0])

print(procedures)
```

```python
#delays per custom_procedure
late = {"DTD": 0, "CRF": 0, "DTP": 0}
total = {"DTD": 0, "CRF": 0, "DTP": 0}

for i, id in enumerate(df_orders["product_id"]):
  total[df_orders[df_orders["product_id"] == id]["customs_procedures"].values[0]] += 1
  if df_orders[df_orders["product_id"] == id]["late_order"].values[0] == True:
    late[df_orders[df_orders["product_id"] == id]["customs_procedures"].values[0]] += 1

for i in late.keys():
    print(i, ":",late[i]/total[i]*100 , "%")
```

```python
# Code you have previously used to load data
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor

# Create target object and call it y
y = df_orders.late_order
# Create X
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = df_orders[features]

# Split into validation and training data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)

# Specify Model
model = DecisionTreeRegressor(random_state=1)
# Fit Model
model.fit(train_X, train_y)

# Make validation predictions and calculate mean absolute error
val_predictions = model.predict(val_X)
val_mae = mean_absolute_error(val_predictions, val_y)
print("Validation MAE when not specifying max_leaf_nodes: {:,.0f}".format(val_mae))

# Using best value for max_leaf_nodes
model = DecisionTreeRegressor(max_leaf_nodes=100, random_state=1)
model.fit(train_X, train_y)
val_predictions = model.predict(val_X)
val_mae = mean_absolute_error(val_predictions, val_y)
print("Validation MAE for best value of max_leaf_nodes: {:,.0f}".format(val_mae))
```
